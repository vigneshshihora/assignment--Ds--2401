```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Find all tables on the page
tables = soup.find_all("table", class_="wikitable")

videos_data = []

for table in tables:
    # Check if the table contains the header row
    if table.find("tr"):
        rows = table.find_all("tr")[1:]  # Exclude header row
        for row in rows:
            columns = row.find_all("td")
            if len(columns) == 5:  # Ensure the row has the expected number of columns
                rank = columns[0].text.strip()
                name = columns[1].text.strip()
                artist = columns[2].text.strip()
                upload_date = columns[3].text.strip()
                views = columns[4].text.strip()
                
                videos_data.append({
                    "Rank": rank,
                    "Name": name,
                    "Artist": artist,
                    "Upload Date": upload_date,
                    "Views": views
                })

# Print the scraped data
for video in videos_data:
    print("Rank:", video["Rank"])
    print("Name:", video["Name"])
    print("Artist:", video["Artist"])
    print("Upload Date:", video["Upload Date"])
    print("Views:", video["Views"])
    print()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# Initialize Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run headless browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Open the BCCI homepage
driver.get("https://www.bcci.tv/")

# Wait for the homepage to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "background")))

# Click on the "International" dropdown menu
international_dropdown = driver.find_element(By.XPATH, "//a[@class='navigation__link dropdown-toggle ng-scope']")
international_dropdown.click()

# Click on the "Fixtures" link
fixtures_link = driver.find_element(By.XPATH, "//a[contains(@class,'navigation__sub-link ng-scope') and text()='Fixtures']")
fixtures_link.click()

# Wait for the fixtures to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "js-list")))

# Find all the fixtures
fixtures = driver.find_elements(By.CLASS_NAME, "js-list")

# Initialize a list to store fixture details
fixture_details = []

# Iterate over each fixture and extract details
for fixture in fixtures:
    series = fixture.find_element(By.CLASS_NAME, "fixture__description").text.strip()
    place = fixture.find_element(By.CLASS_NAME, "fixture__info").text.strip()
    date = fixture.find_element(By.CLASS_NAME, "fixture__date-details").text.strip()
    time = fixture.find_element(By.CLASS_NAME, "fixture__time-details").text.strip()

    fixture_details.append({
        "Series": series,
        "Place": place,
        "Date": date,
        "Time": time
    })

# Print the scraped data
if fixture_details:
    for index, fixture in enumerate(fixture_details, start=1):
        print("A) Series:", fixture["Series"])
        print("B) Place:", fixture["Place"])
        print("C) Date:", fixture["Date"])
        print("D) Time:", fixture["Time"])
        print()
else:
    print("No international fixtures found.")

# Quit the WebDriver
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the website
url = "http://statisticstimes.com/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the link to the economy page
economy_link = soup.find("a", href="/economy")

# Check if the link is found
if economy_link:
    # Construct the URL for the economy page
    economy_url = "http://statisticstimes.com" + economy_link["href"]

    # Send a GET request to the economy page
    economy_response = requests.get(economy_url)

    # Parse the HTML content of the economy page
    economy_soup = BeautifulSoup(economy_response.content, "html.parser")

    # Find the link to the GDP of Indian states page
    gdp_link = economy_soup.find("a", href="/economy/gdp-of-indian-states")

    # Check if the link is found
    if gdp_link:
        # Construct the URL for the GDP of Indian states page
        gdp_url = "http://statisticstimes.com" + gdp_link["href"]

        # Send a GET request to the GDP of Indian states page
        gdp_response = requests.get(gdp_url)

        # Parse the HTML content of the GDP of Indian states page
        gdp_soup = BeautifulSoup(gdp_response.content, "html.parser")

        # Find the table containing State-wise GDP details
        table = gdp_soup.find("table", class_="display")

        # Extract the required details from the table
        rows = table.find_all("tr")[1:]  # Exclude header row
        for row in rows:
            columns = row.find_all("td")
            rank = columns[0].text.strip()
            state = columns[1].text.strip()
            gsdp_1819 = columns[2].text.strip()
            gsdp_1920 = columns[3].text.strip()
            share_1819 = columns[4].text.strip()
            gdp_billion = columns[5].text.strip()

            # Print the scraped details
            print("A) Rank:", rank)
            print("B) State:", state)
            print("C) GSDP(18-19):", gsdp_1819)
            print("D) GSDP(19-20):", gsdp_1920)
            print("E) Share(18-19):", share_1819)
            print("F) GDP($ billion):", gdp_billion)
            print()
    else:
        print("GDP of Indian states link not found.")
else:
    print("Economy link not found.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the GitHub trending page
url = "https://github.com/trending"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all the repository cards
repository_cards = soup.find_all("article", class_="Box-row")

# Extract the required details from each repository card
for card in repository_cards:
    # Repository title
    title_element = card.find("h1", class_="h3")
    title = title_element.text.strip() if title_element else "Title not found"

    # Repository description
    description_element = card.find("p", class_="col-9")
    description = description_element.text.strip() if description_element else "Description not found"

    # Contributors count
    contributors_element = card.find("a", href=True, title="Contributors")
    contributors_count = contributors_element.text.strip() if contributors_element else "Contributors count not found"

    # Language used
    language_element = card.find("span", itemprop="programmingLanguage")
    language = language_element.text.strip() if language_element else "Language not found"

    # Print the scraped details
    print("A) Repository title:", title)
    print("B) Repository description:", description)
    print("C) Contributors count:", contributors_count)
    print("D) Language used:", language)
    print()
```

```python
import asyncio
import nest_asyncio
from requests_html import AsyncHTMLSession

# Allow running asyncio code in Jupyter Notebook
nest_asyncio.apply()

# Create an Async HTML Session
session = AsyncHTMLSession()

# URL of the Billboard website
url = "https://www.billboard.com/charts/hot-100/"

# Define an async function to fetch data
async def fetch_data():
    # Send a GET request to the website
    response = await session.get(url)

    # Render the JavaScript on the page
    await response.html.arender()

    # Extract the chart entries for the top 100 songs
    chart_entries = response.html.find("li.chart-list__element")

    # Initialize a list to store the details of the top 100 songs
    top_songs = []

    # Iterate through the chart entries to extract song details
    for entry in chart_entries:
        # Extract song name
        song_name = entry.find("span.chart-element__information__song", first=True).text

        # Extract artist name
        artist_name = entry.find("span.chart-element__information__artist", first=True).text

        # Extract last week rank
        last_week_rank = entry.find("span.chart-element__meta.text--center.color--secondary.text--last", first=True).text

        # Extract peak rank
        peak_rank = entry.find("span.chart-element__meta.text--center.color--secondary.text--peak", first=True).text

        # Extract weeks on board
        weeks_on_board = entry.find("span.chart-element__meta.text--center.color--secondary.text--week", first=True).text

        # Append song details to the list
        top_songs.append({
            "Song Name": song_name,
            "Artist Name": artist_name,
            "Last Week Rank": last_week_rank,
            "Peak Rank": peak_rank,
            "Weeks on Board": weeks_on_board
        })

    # Print the details of the top 100 songs
    for i, song in enumerate(top_songs, start=1):
        print(f"{i}.")
        print("A) Song Name:", song["Song Name"])
        print("B) Artist Name:", song["Artist Name"])
        print("C) Last Week Rank:", song["Last Week Rank"])
        print("D) Peak Rank:", song["Peak Rank"])
        print("E) Weeks on Board:", song["Weeks on Board"])
        print()

# Run the async function directly
await fetch_data()
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the highest selling novels
novels_table = soup.find("table", class_="in-article sortable")

# Check if the table is found
if novels_table:
    # Initialize a list to store the details of the novels
    novels_list = []

    # Extract details for each novel
    for row in novels_table.find_all("tr")[1:]:  # Skip the header row
        columns = row.find_all("td")
        
        # Add error handling in case a row doesn't have enough columns
        if len(columns) >= 5:
            book_name = columns[1].text.strip()
            author_name = columns[2].text.strip()
            volumes_sold = columns[3].text.strip()
            publisher = columns[4].text.strip()
            genre = columns[5].text.strip()

            # Append the details to the novels_list
            novels_list.append({
                "Book Name": book_name,
                "Author Name": author_name,
                "Volumes Sold": volumes_sold,
                "Publisher": publisher,
                "Genre": genre
            })
        else:
            print("Row doesn't have enough columns:", [col.text.strip() for col in columns])

    # Print the details of the highest selling novels
    for novel in novels_list:
        print("Book Name:", novel["Book Name"])
        print("Author Name:", novel["Author Name"])
        print("Volumes Sold:", novel["Volumes Sold"])
        print("Publisher:", novel["Publisher"])
        print("Genre:", novel["Genre"])
        print()
else:
    print("The section containing information about highest selling novels couldn't be found.")
```
